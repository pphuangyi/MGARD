# Compress TPC data with MGARD

```python
import numpy as np 
import pandas as pd
import re
from subprocess import check_output
from pathlib import Path
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Functions and Global Parameters

### Parse values from output string

```python
def get_value(content, key):
    assert content.find(key) != -1, \
        f"Content doesn't contain key '{key}'"
    key_ = f'{key}: '
    assert content.find(key_) != -1, \
        f"Content does contain the key '{key}', but the format is wrong"
    
    key_length = len(key_)
    start = content.find(key_) + key_length
    first_delimiter = re.search('\s+', content[start : ]).start()
    value = content[start : start + first_delimiter]
    
    try:
        value = float(value)
        return value
    except:
        print(f"Content doesn't contains a value for {key}.")
```

### Global parameters

```python
data_folder = '../tpc_data/'
shape_map = {
    'outer': [192, 249, 16],
    'middle': [128, 249, 16],
    'inner': [96, 249, 16]
}
program = '../build/tpc'
save_folder = '../tpc_results'
```

### Run MGARD on TPC data

```python
def run_mgard_tpc(layer_group, l_infinity_bound, max_examples=None):
    fnames = [path for path in Path(data_folder).glob(f'*{layer_group}*dat')]
    if max_examples and len(fnames) > max_examples:
        fnames = np.random.choice(fnames, max_examples, replace=False)
    
    shape_str = ' '.join(map(str, shape_map[layer_group]))

    results = []
    for i, fname in enumerate(fnames):
        cmd = f"srun {program} {fname} {l_infinity_bound} 0 {shape_str}"
        output = check_output(cmd, shell=True).decode("utf-8")

        mse = get_value(output, 'MSE')
        nonzero_count = get_value(output, 'Non-zero count')
        nonzero_mse = get_value(output, 'Non-zero MSE')
        ratio = get_value(output, 'Compression ratio')
        l_infinity_error = get_value(output, 'Abs. L-infinity error') 

        print(f'{i + 1} / {len(fnames)}: {fname.name}')
        print(f'\tmse = {mse}')
        print(f'\tnonzero count = {nonzero_count}')
        print(f'\tnonzero mse = {nonzero_mse}')
        print(f'\tratio = {ratio}')
        print(f'\tl infinity error = {l_infinity_error}\n')
        result = [fname.name, mse, nonzero_count, nonzero_mse, ratio, l_infinity_error]
        results.append(result)
    columns = ['file_name', 'mse', 'nonzero_count', 'nonzero_mse', 'compression_ratio', 'l_infinity_error']
    df = pd.DataFrame(data=results, columns=columns)
    df['l_infinity_bound'] = l_infinity_bound
    # df['inv_compression'] = round(1 / df['compression_ratio'])
    return df
```

## Run

```python
l_infinity_bounds = np.arange(20, 520, 20)
layer_group = 'outer'
```

```python
for l_infinity_bound in l_infinity_bounds:
    print(f'\n======================= {l_infinity_bound} ======================')
    df = run_mgard_tpc(layer_group, l_infinity_bound, max_examples=100)
    dataframe_name = f'{save_folder}/mgard_{layer_group}_{l_infinity_bound}.csv'
    dfs.append(df)
    df.to_csv(dataframe_name, index=False)
```

```python
dfs = []
for l_infinity_bound in l_infinity_bounds:
    df = pd.read_csv(f'{save_folder}/mgard_group_lbound/mgard_{layer_group}_{l_infinity_bound}.csv')
    dfs.append(df)
df_combined = pd.concat(dfs, axis=0).reset_index(drop=True)
df_combined.to_csv(f'{save_folder}/mgard_{layer_group}.csv', index=False)
```

## Plot

```python
benchMark_line = { 'outer': 27, 'middle': 13, 'inner': 13 }
for layer_group in ['outer', 'middle', 'inner']:
    df = pd.read_csv(f'{save_folder}/mgard_{layer_group}.csv').drop('file_name', axis=1)
    df_avg = df.groupby(by='l_infinity_bound').mean()
    df_avg['inv_compression_ratio'] = 1 / df_avg['compression_ratio']
    df_avg.sort_values(by='inv_compression_ratio', inplace=True)

    fig, ax = plt.subplots(1, 1, figsize=(8, 3))
    ax.plot(df_avg['inv_compression_ratio'], df_avg['mse'], marker='o', label='mse')
    ax.plot(df_avg['inv_compression_ratio'], df_avg['nonzero_mse'], marker='o', label='non-zero mse')
    ax.set_xlabel('Compression ratio', fontsize=16)
    ax.set_ylabel('MSE', fontsize=16)
    ax.axvline(benchMark_line[layer_group], color='k')
    ax.set_title(f'{layer_group}', fontsize=20)
    ax.grid()
    ax.legend()
    fig.savefig(f'{save_folder}/mse-to-cr_{layer_group}.png')
```
